```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import csv
```

```python
html = urlopen('https://www.oliveyoung.co.kr/store/main/getBestList.do')
bsoup = BeautifulSoup(html, "html.parser")
```

```python
bsoup
```

```python
a = bsoup.findAll("div", {"class":"prd_info"})
```

```python
for bf in a:
    rank = bf.find("span", {"class":"thumb_flag best"}).get_text()
    title = bf.find("p", {"class":"tx_name"}).get_text()
    brand = bf.find("span",{"class":"tx_brand"}).get_text()
    cost = bf.find("span", {"class":"tx_num"}).get_text()
    costnow = bf.find("span", {"class":"tx_cur"}).get_text()
    print(rank, title, brand, cost, costnow)
```

```python
with open('oliveyoungcrwaling.csv', 'w') as f:
    writer = csv.writer(f)
    writer.writerow(['순위', '이름', '브랜드', '세일전', '세일후'])
        
    for bf in a:
        rank = bf.find("span", {"class":"thumb_flag best"}).get_text()
        title = bf.find("p", {"class":"tx_name"}).get_text()
        brand = bf.find("span",{"class":"tx_brand"}).get_text()
        cost = bf.find("span", {"class":"tx_num"}).get_text()
        costnow = bf.find("span", {"class":"tx_cur"}).get_text()
        writer.writerow([rank, title, brand, cost, costnow])
```
