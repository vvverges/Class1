def부터 return까지의 3개 줄이 작성한 함수
설정한 regexp : bar를 기준으로 세등분 되어있다. 
    bar를 기준으로 첫번째 조건 : ^[AEIOUaeiou]+ : ^(string의 처음에서) 모음으로 시작되는 것이 하나 이상 나오면 찾아라
                   두번째 조건 : 마지막에서 모음인 것을 찾아라
                   세번째 조건 : 자음인 것을 찾아라
def : define
함수의 이름 : compress ()는 입력되는 값
함수의 내용은 tab으로 indent를 띄어 주어야 한다.
compress라는 함수는 word라는 string을 받아서 앞에서 설정해 둔 regexp를 적용
return은 출력값을 의미한다. : join의 입력은 list들이고, ''(glue)로 붙인다. (space 공백 없이)

-> 처음과 끝에 나오는 모음만 모두 살리고 나머지는 자음만 살린다
ex) universal -> unvrsl

```python
regexp = r'^[AEIOUaeiou]+|[AEIOUaeiou]+$|[^AEIOUaeiou]'
def compress(word):
    pieces = re.findall(regexp, word)
    return ''.join(pieces)

english_udhr = nltk.corpus.udhr.words('English-Latin1')
print(nltk.tokenwrap(compress(w) for w in english_udhr[:75]))
```

ConditionalFreqDist : pairwise인 경우에만 쓰인다.
tabulate한다.
number of occurrencies가 conditionally하게 도출된다.

```python
rotokas_words = nltk.corpus.toolbox.words('rotokas.dic')
cvs = [cv for w in rotokas_words for cv in re.findall(r'[ptksvr][aeiou]', w)]
cfd = nltk.ConditionalFreqDist(cvs) # CFD works pairwise only
cfd.tabulate()
```
